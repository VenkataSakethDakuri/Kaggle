```python
train_path = '/kaggle/input/titanic/train.csv'
train_data = pd.read_csv(train_path)
train_data.head()

test_path = '/kaggle/input/titanic/test.csv'
test_data = pd.read_csv(test_path)
```

```python
print(len(train_data))
print(train_data.columns)
```

```python
women_survived = train_data.loc[train_data.Sex == 'female'][train_data.Survived == 1]
total_women = train_data.loc[train_data.Sex == 'female']
percent_of_women_survived = (len(women_survived) / len(total_women)) * 100
print("% of women survived", percent_of_women_survived)
```

```python
missing_values_embarked = train_data['Embarked'].isna().sum()
print(missing_values_embarked)

missing_values_cabin = train_data['Cabin'].isna().sum()
print(missing_values_cabin)

train_data = train_data.drop(columns = ['Cabin'])
```

```python
features = ["Pclass", "Sex",  "SibSp", "Parch"]
# get_dummmies ignores columns with numerical values
#get_dummies handles missing values by keeping zeroes 
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

y = train_data['Survived']

#make sure the same columns exist in both the train and test set
X, X_test = X.align(X_test, join = 'inner', axis = 1)

print(X.shape[1])
print(X_test.shape[1])
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 100, max_depth = 5, random_state = 108)
model.fit(X, y)

predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index = False)
```
